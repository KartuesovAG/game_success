### Описание данных
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

### Предобработка данных

```python
#привожу название столбцов к нижнему регистру
df.columns = df.columns.str.lower()
df.head()
```

```python
#проверяю пропуски в датасете
df.isna().sum()
```

```python
#проверяю два пропуска в name / genre 
df[df['name'].isna()]
```

Пропуски в name / genre идентичны, можно удалить. Скорее всего технический сбой.

```python
df = df[~df['name'].isna()]
```

```python
#проверяю пропуски в годе выпуска
df[df['year_of_release'].isna()]
```

В некоторых названиях игр указан год выпуска, но мне кажется, что эти строчки легче удалить и они не повлияют на дальнейшие исследования. Возможная причина ошибки - человеческий фактор, так как в некоторых данных указана дата выпуска игры.

```python
df = df[~df['year_of_release'].isna()]
```

```python
#изменение типа данных год выпуска
df['year_of_release'] = df['year_of_release'].astype(int)
```

```python
#необходимо рассмотреть строки со значением tbd
df[df['user_score'] == 'tbd'].head()
```

tbd - to be determinaed, поэтому их можно приравнять к категории пропуска с использованием метода to_numeric библиотеки pandas

```python
#изменение типа данных оценки пользователей
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')
```

```python
#проверка датасета на явные дубликаты
df.duplicated().sum()
```

```python
#проверка на неявные дубликаты
df['genre'].unique()
```

```python
df['rating'] = df['rating'].fillna('Unknown')
```

```python
df['platform'].unique()
```

Неявные дубликатов нет

```python
#суммарные продажи по всем регионам
df['total_sales'] = (df['na_sales']\
                    + df['eu_sales']\
                    + df['jp_sales']\
                    + df['other_sales']
                    )
```

```python
temp = df.copy() 
list_c = ['name', 'platform', 'year_of_release', 'genre', 'critic_score', 'user_score', 'rating']
print(temp.info())
for col_l in list_c:
    print('-'* 25)
    print(col_l, temp[col_l].sort_values().unique())
    print(col_l,': кол-во NaN',temp[col_l].isna().sum(),
        ', процент NaN', round(temp[col_l].isna().sum()/len(temp)*100, 2),'%')
```

### Исследовательский анализ данных

#### Сколько игр выпускалось в разные годы?

```python
#создаю сводную таблицу по количеству релизов игр в разные годы и строю график
game_release = df.pivot_table(index='year_of_release', values='name', aggfunc='count').reset_index()
game_release.columns = ['Год', 'Количество игр']
display(game_release.sort_values(by='Количество игр', ascending = False).reset_index(drop=True).head(15))
game_release.plot(x='Год',
                  grid=True,
                  xlabel='Год',
                  ylabel='Количество',
                  label='Количество игр',
                  title='Зависимость выпуска игр в разные годы',
                  style='o-'
                 );
```

На графике видно, что начиная с 1994 начинается значительный рост выпуска игр. Пик выпуска игр приходится на 2008 и 2009 год. 

Для дальнейшей работы предварительно буду использовать данные за последние 10 лет, то есть от 2006 года (включительно).

```python
#создам новую переменную с данными от 2006 года, исходный датасет лучше оставить без изменений
ddf = df[df['year_of_release'] >= 2006]
```

#### Как менялись продажи по платформам?

```python
platform_sales = (ddf
                    .pivot_table(index='platform', values ='total_sales', aggfunc='sum')
                    .sort_values(by='total_sales', ascending=False)
                    .reset_index()
                 )

platform_sales.head(10)
```

В таблице отобразил ТОП-10 платформ по продажам

```python
#создаю новую сводную таблицу для построения распределения по годам
platform_year_sales = (ddf
                       .pivot_table(index=['platform', 'year_of_release'],
                                    values ='total_sales',
                                    aggfunc='sum')
                       .sort_values(by='total_sales',ascending=False)
                       .reset_index()
                      )
platform_year_sales
```

```python
#чтобы не выводить вручную делаю переменную с данными ТОП-10 платформ по продажам
top_10_sales = platform_sales.loc[:9, 'platform'].tolist()
#через цикл строю графики для каждой из ТОП-10 платформ
for i in top_10_sales:
        (platform_year_sales.loc[platform_year_sales['platform'] == i]
         .reset_index()
         .sort_values(by='year_of_release')
         .plot(x='year_of_release',
               y='total_sales',
               grid=True,
               xlabel='Год',
               ylabel='Количество продаж',
               label=i,
               figsize=(5, 5),
               kind='bar')
        )
print('ТОП-10 платформ по количеству продаж')
```

На гистограммах видно, что в среднем платформы актуальны на течении 6 лет, после чего выходит новая ее версия. 

Актуальным периодом для исследования можно считать данные с 2013 по 2016 год.


```python
#делаю срез данных для актуального периода
actual_df = df.query('2013 <= year_of_release <= 2016')
```

#### Какие платформы лидируют по продажам?

```python
actual_sales = (actual_df 
                .pivot_table(index='platform', values='total_sales', aggfunc='sum')
                .sort_values(by='total_sales', ascending = False)
                .reset_index())
actual_sales.plot(kind='bar',
                      x='platform',
                      y='total_sales',
                      grid=True,
                      xlabel='Платформа',
                      ylabel='Количество продаж',
                      label='Продажи',
                      title='Суммарные продажи по платформам за период 2013-2016гг.');
```

На гистограмме видно, что наиболее прибыльными платформами являются: PS4, PS3, XOne, 3DS и X360.

#### Описание глобальных продаж игр в разбивке по платформам. 

```python
display(actual_df.groupby('platform')['total_sales'].describe().T)
actual_df.boxplot(column='total_sales', by='platform',showfliers=True); 
plt.xlabel('Платформа') 
plt.ylabel("Суммарные продажи, млн. копий")
plt.title('''Диаграмма размаха глобальных продаж по платформам''')
;
actual_df.boxplot(column='total_sales', by='platform',showfliers=False)
plt.xlabel('Платформа') 
plt.ylabel("Суммарные продажи")
plt.title('''Диаграмма размаха глобальных продаж по платформам''')
```

На диаграмме размаха видно, что медианные продажи всех платформ (исключение X360) лежат в одном диапозоне - до 250 тысяч копий. Однако, продажи выше третьего квартиля особенно заметны среди платформ: PS4, X360, Wii, XOne  и WiiU. 

#### Как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков? 

```python
#создаю новую переменную для данных по платформе PS4
df_ps4 = actual_df[actual_df['platform'] == 'PS4'][['total_sales', 'critic_score', 'user_score']]
df_ps4.plot(x='total_sales',
            y='critic_score',
            kind='scatter',
            alpha=0.4,
            title='Зависимость продаж игр на платформе PS4 от оценки критиков',
            xlabel='Продажи игр',
            ylabel='Оценка критиков');
df_ps4.plot(x='total_sales',
            y='user_score',
            kind='scatter',
            alpha=0.4,
            title='Зависимость продаж игр на платформе PS4 от оценки пользователей',
            xlabel='Продажи игр',
            ylabel='Оценка пользователей');
print(df_ps4.corr())
```

- Коэффициент корреляции продаж игр на платформе PS4 от оценки критиков положительный, связь слабая.
- Коэффициент корреляции продаж игр на платформе PS4 от оценки пользователей отрицательный, связь отсутствует.
- Зависимость продаж видеоигр с оценками критиков и пользователей отсутствует.

```python
platform_corr = ['PS3', 'X360', '3DS', 'XOne', 'WiiU']
for i in platform_corr:
    
    corr = actual_df[actual_df['platform'] == i][['total_sales', 'critic_score', 'user_score']].corr()
    print('Корреляция продаж платформы {0} от оценки критиков и пользователей'.format(i))
    display(corr)
    actual_df[actual_df['platform'] == i].plot(x='total_sales',
                                               y='critic_score',
                                               kind='scatter',
                                               label=i,
                                               alpha=0.4,
                                               xlabel='Продажи игр',
                                               ylabel='Оценка критиков')
    plt.title('Зависимость продаж платформы {0} от оценки критиков'.format(i))
    plt.show()
    actual_df[actual_df['platform'] == i].plot(x='total_sales',
                                               y='user_score',
                                               kind='scatter',
                                               label=i,
                                               alpha=0.4,
                                               xlabel='Продажи игр',
                                               ylabel='Оценка критиков')
    plt.title('Зависимость продаж платформы {0} от оценки пользователей'.format(i))
    plt.show()
```

### Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```python
XOne = actual_df[(actual_df['platform'] == 'XOne')]['user_score']
PC = actual_df[(actual_df['platform'] == 'PC')]['user_score']
print('Средний пользовательский рейтинг платформы Xbox One:', XOne.mean())
print('Средний пользовательский рейтинг платформы PC:', PC.mean())
```

Гипотезы:

H0 - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H1 - Средние пользовательские рейтинги платформ Xbox One и PC разные

```python
#удаление пропусков
df_xone = actual_df.loc[actual_df['platform'] == 'XOne', 'user_score'].dropna()
df_pc = actual_df.loc[actual_df['platform'] == 'PC', 'user_score'].dropna()
```

Две выборки отличаются незначительно по количеству значений.

```python
alpha = 0.05 #критический уровень стат. значимости

#необходимо удалить строки с пропусками в user_score
results = st.ttest_ind(df_xone, df_pc, equal_var=False)

print('p-значение равно:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Не получилось отвергнуть нулевую гипотезу, что средние пользовательские рейтинги платформ Xbox One и PC одинаковы.

#### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

```python
action = actual_df[(actual_df['genre'] == 'Action') & (actual_df['user_score'] > 0)]['user_score']
sports = actual_df[(actual_df['genre'] == 'Sports') & (actual_df['user_score'] > 0)]['user_score']
print('Средний пользовательский рейтинг жанра Action:', action.mean())
print('Средний пользовательский рейтинг жанра Sports:', sports.mean())
```

Гипотезы:

H0 - Средние пользовательские рейтинги жанров Action и Sports одинаковые

H1 - Средние пользовательские рейтинги жанров Action и Sports разные

```python
#удаляю пропуски
df_action = actual_df.loc[actual_df['genre'] == 'Action', 'user_score'].dropna()
df_sports = actual_df.loc[actual_df['genre'] == 'Sports', 'user_score'].dropna()
```

```python
alpha = 0.05 #критический уровень стат. значимости

#необходимо удалить строки с пропусками в user_score
results = st.ttest_ind(df_action, df_sports, equal_var=False)

print('p-значение равно:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Отвергаем нулевую гипотезу, что средние пользовательские рейтинги жанров Action и Sports одинаковые.
